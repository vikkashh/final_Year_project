```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("DF1_FullSignal.xlsx")
```

```python
print(df.shape)
print(df.columns)

df.info()
df.describe()
```

```python
df1=df
```

```python
print(df1.shape)
print(df1.info())
```

```python
print(df1.isnull().sum().sum())
```

```python
df1["Condition"].value_counts().plot(kind="bar")
plt.title("Condition Distribution")
plt.show()
```

```python
df1["Amplitude"].value_counts().plot(kind="bar")
plt.title("Amplitude Distribution")
plt.show()
```

```python
sns.boxplot(x="Condition", y="S1_rms", data=df1)
plt.title("RMS vs Condition")
plt.show()
```

```python
sns.boxplot(x="Amplitude", y="S1_rms", data=df1)
plt.title("Amplitude vs RMS")
plt.show()
```

```python
sns.boxplot(x="Level", y="S1_rms", data=df1)
plt.title("Level vs RMS")
plt.show()
```

```python
df1["S1_rms"].hist(bins=50)
plt.title("RMS Distribution")
plt.show()
```
