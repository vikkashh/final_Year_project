```python
# 🔹 Basic Libraries
import pandas as pd
import numpy as np

# 🔹 Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# 🔹 Preprocessing
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 🔹 Train-Test Split
from sklearn.model_selection import train_test_split

# 🔹 Models
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# 🔹 Boosting Models
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# 🔹 Evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 🔹 Hyperparameter Tuning
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
# Load Excel file
df = pd.read_excel("DF4_Mag_Windowed.xlsx")

# View first rows
df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe()
```

```python
df.drop(columns=["FileName", "Window"],inplace=True)
```

```python
df.head()
```

```python
sns.heatmap(df.drop(columns=["Condition"]).corr())
```

```python
def create_target(df):
    df["Target"] = df["Condition"] + "_L" + df["Level"].astype(str)
    return df
```

```python
df=create_target(df)
```

```python
# Drop unnecessary columns
X = df.drop(columns=["Condition", "Level", "Target"])

# Target
y = df["Target"]

print("Features shape:", X.shape)
```

```python
train_columns = X.columns
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("Classes:", le.classes_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report
)
```

```python


def evaluate_model(model, X_test, y_test, model_name="Model"):
    
    # Prediction
    y_pred = model.predict(X_test)
    
    # Metrics
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    print(f"\n🔹 {model_name}")
    print("Accuracy :", acc)
    print("Precision:", precision)
    print("Recall   :", recall)
    print("F1 Score :", f1)
    
    print("\nClassification Report:\n")
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    return acc, precision, recall, f1
```

```python

from sklearn.linear_model import LogisticRegression

best_score1 = 0
best_params1 = None

C_values = [0.01, 0.1, 1, 10,20,50,100]
penalties = ['l2']   # l1 requires different solver

results = []

for c in C_values:
    for p in penalties:
        
        model = LogisticRegression(
            C=c,
            penalty=p,
            solver='lbfgs',   # supports l2
            max_iter=1000
        )
        
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        
        acc = accuracy_score(y_test, y_pred)

        if acc > best_score1:
            best_score1 = acc
            best_params1 = c
        
        results.append([c, p, acc])

df_lr_plot = pd.DataFrame(results, columns=["C", "penalty", "accuracy"])

print(df_lr_plot)
```

```python
pivot = df_lr_plot.pivot(index="penalty", columns="C", values="accuracy")

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, cmap="viridis", fmt=".4f")

plt.title("Logistic Regression: C vs penalty")
plt.xlabel("C")
plt.ylabel("penalty")

plt.show()
```

```python
plt.figure(figsize=(8,6))

for p in penalties:
    subset = df_lr_plot[df_lr_plot["penalty"] == p]
    
    plt.plot(
        subset["C"],
        subset["accuracy"],
        marker='o',
        label=f"penalty={p}"
    )

plt.title("C vs Accuracy (for penalty)")
plt.xlabel("C")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
```

```python
lr_final = LogisticRegression(C=best_params1, penalty='l2', solver='lbfgs',max_iter=1000)
```

```python
lr_final.fit(X_train_scaled, y_train)
evaluate_model(lr_final, X_test_scaled, y_test, "Logistic Regression")
```

```python

C_values = [0.1, 1, 10, 50]
gamma_values = ['scale', 'auto']

results = []
best_score2 = 0
best_params2 = None
for c in C_values:
    for g in gamma_values:
        
        model = SVC(C=c, gamma=g, kernel='rbf')
        model.fit(X_train_scaled, y_train)
        
        y_pred = model.predict(X_test_scaled)
        acc = accuracy_score(y_test, y_pred)

        if acc > best_score2:
            best_score2 = acc
            best_params2 =[c, g]
        results.append([c, g, acc])

df_svm_plot = pd.DataFrame(results, columns=["C", "gamma", "accuracy"])
```

```python
pivot = df_svm_plot.pivot(index="gamma", columns="C", values="accuracy")

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, cmap="viridis", fmt=".4f")

plt.title("SVM: C vs Gamma")
plt.xlabel("C")
plt.ylabel("Gamma")

plt.show()
```

```python
plt.figure(figsize=(8,6))

for g in gamma_values:
    subset = df_svm_plot[df_svm_plot["gamma"] == g]
    
    plt.plot(
        subset["C"],
        subset["accuracy"],
        marker='o',
        label=f"gamma={g}"
    )

plt.title("C vs Accuracy (for different gamma)")
plt.xlabel("C")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,6))

for c in C_values:
    subset = df_svm_plot[df_svm_plot["C"] == c]
    
    plt.plot(
        subset["gamma"],
        subset["accuracy"],
        marker='o',
        label=f"C={c}"
    )

plt.title("Gamma vs Accuracy (for different C)")
plt.xlabel("Gamma")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
```

```python
gamma_map = {'scale': 0, 'auto': 1}
df_svm_plot["gamma_num"] = df_svm_plot["gamma"].map(gamma_map)
```

```python
from mpl_toolkits.mplot3d import Axes3D

C_vals = sorted(df_svm_plot["C"].unique())
G_vals = sorted(df_svm_plot["gamma_num"].unique())

C_grid, G_grid = np.meshgrid(C_vals, G_vals)
Z = np.zeros_like(C_grid, dtype=float)

for i in range(len(G_vals)):
    for j in range(len(C_vals)):
        val = df_svm_plot[
            (df_svm_plot["C"] == C_grid[i, j]) &
            (df_svm_plot["gamma_num"] == G_grid[i, j])
        ]["accuracy"].values
        
        if len(val) > 0:
            Z[i, j] = val[0]

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(C_grid, G_grid, Z, cmap='viridis')

ax.set_xlabel("C")
ax.set_ylabel("Gamma (0=scale,1=auto)")
ax.set_zlabel("Accuracy")

plt.title("SVM 3D Surface")

plt.show()
```

```python
import pandas as pd
import numpy as np

depth_values = [5, 10, 20, None]
split_values = [2, 5, 10]

results = []
best_score3 = 0
best_params3 = None

for d in depth_values:
    for s in split_values:
        
        model = DecisionTreeClassifier(
            max_depth=d,
            min_samples_split=s,
            random_state=42
        )
       
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        
        acc = accuracy_score(y_test, y_pred)

        if acc > best_score3:
            best_score3 = acc
            best_params3 =[d, s]
        results.append([d, s, acc])

df_dt_plot = pd.DataFrame(results, columns=["max_depth", "min_samples_split", "accuracy"])
```

```python
plt.figure(figsize=(8,6))

for d in df_dt_plot["max_depth"].unique():
    subset = df_dt_plot[df_dt_plot["max_depth"] == d]
    
    plt.plot(
        subset["min_samples_split"],
        subset["accuracy"],
        marker='o',
        label=f"max_depth={d}"
    )

plt.title("min_samples_split vs Accuracy (for different max_depth)")
plt.xlabel("min_samples_split")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

n_values = [100, 200, 300]
depth_values = [5, 10, 20]
best_score4 = 0
best_params4 = None
results = []

for n in n_values:
    for d in depth_values:
        model = RandomForestClassifier(
            n_estimators=n,
            max_depth=d,
            random_state=42
        )
        
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        
        acc = accuracy_score(y_test, y_pred)

        if acc > best_score4:
            best_score4 = acc
            best_params4 = [n, d]
            
        results.append([n, d, acc])

df_plot = pd.DataFrame(results, columns=["n_estimators", "max_depth", "accuracy"])
```

```python
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# 🔥 Reduced but effective values
n_values = [100, 200,300]
lr_values = [0.05, 0.1,0.5,1]

results = []
best_score5 = 0
best_params5 = None

for n in n_values:
    for lr in lr_values:
        
        model = XGBClassifier(
            n_estimators=n,
            learning_rate=lr,
            max_depth=5,        # good default
            subsample=0.8,      # faster + prevents overfitting
            colsample_bytree=0.8,
            eval_metric='mlogloss',
            n_jobs=-1           # 🔥 parallel processing
        )
       
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        
        acc = accuracy_score(y_test, y_pred)

        if acc > best_score5:
            best_score5 = acc
            best_params5 = [n, lr]
        results.append([n, lr, acc])

df_xgb_plot = pd.DataFrame(results, columns=["n_estimators", "learning_rate", "accuracy"])

print(df_xgb_plot)
```

```python
pivot = df_xgb_plot.pivot(index="learning_rate", columns="n_estimators", values="accuracy")

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, cmap="viridis", fmt=".4f")

plt.title("XGBoost: n_estimators vs learning_rate")
plt.xlabel("n_estimators")
plt.ylabel("learning_rate")

plt.show()
```

```python
plt.figure(figsize=(8,6))

for lr in sorted(df_xgb_plot["learning_rate"].unique()):
    subset = df_xgb_plot[df_xgb_plot["learning_rate"] == lr]
    
    plt.plot(
        subset["n_estimators"],
        subset["accuracy"],
        marker='o',
        label=f"lr={lr}"
    )

plt.title("n_estimators vs Accuracy (for different learning_rate)")
plt.xlabel("n_estimators")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,6))

for n in sorted(df_xgb_plot["n_estimators"].unique()):
    subset = df_xgb_plot[df_xgb_plot["n_estimators"] == n]
    
    plt.plot(
        subset["learning_rate"],
        subset["accuracy"],
        marker='o',
        label=f"n={n}"
    )

plt.title("learning_rate vs Accuracy (for different n_estimators)")
plt.xlabel("learning_rate")
plt.ylabel("Accuracy")
plt.legend()
plt.grid()

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

N_vals = sorted(df_xgb_plot["n_estimators"].unique())
LR_vals = sorted(df_xgb_plot["learning_rate"].unique())

N_grid, LR_grid = np.meshgrid(N_vals, LR_vals)
Z = np.zeros_like(N_grid, dtype=float)

for i in range(len(LR_vals)):
    for j in range(len(N_vals)):
        val = df_xgb_plot[
            (df_xgb_plot["n_estimators"] == N_grid[i, j]) &
            (df_xgb_plot["learning_rate"] == LR_grid[i, j])
        ]["accuracy"].values
        
        if len(val) > 0:
            Z[i, j] = val[0]

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(N_grid, LR_grid, Z, cmap='viridis')

ax.set_xlabel("n_estimators")
ax.set_ylabel("learning_rate")
ax.set_zlabel("Accuracy")

plt.title("XGBoost 3D Surface")

plt.show()
```
