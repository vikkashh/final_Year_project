```python
import joblib
import pandas as pd
import numpy as np

# Load models
svm = joblib.load("svm.pkl")
lgb = joblib.load("lgb.pkl")
rf = joblib.load("rf.pkl")
xgb = joblib.load("xgb.pkl")

# Load preprocessing
scaler = joblib.load("scaler.pkl")
le = joblib.load("label_encoder.pkl")
x_columns = joblib.load("columns.pkl")
from collections import Counter
```

```python
def create_df4_from_df(df, amplitude, condition, level, windows=30):
    
    import numpy as np
    import pandas as pd

    # -------- FEATURE FUNCTION --------
    def extract_features(arr):
        return {
            "min": np.min(arr),
            "max": np.max(arr),
            "mean": np.mean(arr),
            "std": np.std(arr),
            "var": np.var(arr),
            "rms": np.sqrt(np.mean(arr**2)),
            "mav": np.mean(np.abs(arr))
        }

    # -------- RAW DATA (24 signals) --------
    data = df.iloc[:, :24].values

    # -------- MAGNITUDE (8 sensors) --------
    magnitude = []

    for i in range(0, 24, 3):
        x = data[:, i]
        y = data[:, i + 1]
        z = data[:, i + 2]
        mag = np.sqrt(x**2 + y**2 + z**2)
        magnitude.append(mag)

    magnitude = np.array(magnitude).T  # (N, 8)

    # -------- WINDOWING --------
    window_size = len(magnitude) // windows
    df4_list = []

    for w in range(windows):
        start = w * window_size
        end = (w + 1) * window_size
        window = magnitude[start:end]

        features = {}

        for i in range(8):
            f = extract_features(window[:, i])
            for k, v in f.items():
                features[f"M{i+1}_{k}"] = v

        # -------- ADD LABEL INFO --------
        features.update({
            "Amplitude": amplitude,
            "Condition": condition,
            "Level": level,
        })

        df4_list.append(features)

    df4 = pd.DataFrame(df4_list)

    return df4
```

```python
df_new = pd.read_csv("experiment_8-1.csv")
```

```python
# ================= INPUT FORMAT =================
# amplitude → 0.5, 1, 2
# condition → "Healthy", "6Nm", "9Nm", "NoBolt"
# level     → 0, 1, 2, 3, 4

# ⚠️ IMPORTANT RULE:
# "Healthy" → level must be 0
# "6Nm", "9Nm", "NoBolt" → level must be 1–4

# ❌ Wrong:
# condition="healthy"  (not allowed)
# condition="6nm"      (not allowed)

# ✅ Correct:
# condition="Healthy"
# condition="6Nm"
# ================================================
```

```python
df4_new =create_df4_from_df(df_new, amplitude=0.5, condition="6Nm", level=1)

print(df4_new.head())
```

```python
X_new = df4_new.drop(columns=["Condition", "Level", "Window"], errors="ignore")

# Match training column order
X_new = X_new[x_columns]

# Scale
X_scaled = scaler.transform(X_new)

# ================= PREDICTION =================
models = [lgb, rf, xgb]

probs = [m.predict_proba(X_scaled) for m in models]
avg_prob = sum(probs) / len(probs)

pred = np.argmax(avg_prob, axis=1)

# Convert numeric → label
final_pred = le.inverse_transform(pred)

# ================= WINDOW RESULT =================
print("Window Predictions:")
print(final_pred)

# ================= FINAL RESULT =================
final_condition = Counter(final_pred).most_common(1)[0][0]

print("\nFinal Predicted Condition:", final_condition)
```
